```python
#checking the null values
# hw to replace null values mean or median?
#mean can get affected by outliers that can screw distribution of our data
#median is robust to outliers
df.isnull().sum()
```

```python
#It fills missing valuesin the "Review Rating" column with the median rating of the same category.
#df.groupby("Category"): Groups the DataFrame by the "Category" column.

#["Review Rating"]: Focuses on the "Review Rating" column within each group.

#transform("median"): Calculates the median of "Review Rating" for each group and returns a Series aligned with the original DataFrame.

#df["Review Rating"].fillna(...): Replaces missing values in "Review Rating" with the corresponding group median.

df["Review Rating"] = df["Review Rating"].fillna(df.groupby("Category")["Review Rating"].transform("median"))
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# create a new column age_group

#pd.qcut(): This function divides the data into quantiles, meaning each bin will have (roughly) the same number of data points.

#q=4: Specifies that the "age" column should be divided into 4 quantile-based bins.

#labels=labels: Assigns custom labels to each bin: 'Young Adult', 'Adult', 'Middle-aged', and 'Senior'
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days
#create a dictionary and map

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
